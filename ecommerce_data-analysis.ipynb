# Monthly Sales Analysis

```python
sales_by_month=df.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
plt.figure(figsize=(10,8))
ax=sns.lineplot(data=sales_by_month, x='Order Month', y='Sales' )
plt.title('Monthly Sales Analysis')
plt.xlabel('Order Month')
plt.ylabel('Sales')


plt.tight_layout()
plt.show()
```

# Sales Analysis by Category

```python
sales_by_category=df.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python

labels = sales_by_category['Category']
sizes = sales_by_category['Sales']
plt.figure(figsize=(4, 8))
plt.pie(sizes, labels=labels, autopct='%0.1f%%',)
plt.title('Sales Analysis by Category')
plt.show()
```

# Conclusion : Technology then Furniture and then  Office Supplies

# Sales Analysis by  Sub-Category 

```python
sales_by_sub_category=df.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_sub_category
```

```python

plt.figure(figsize=(12, 6))
ax=sns.barplot(data=sales_by_sub_category, x='Sub-Category', y='Sales', palette='pastel')
for container in ax.containers:
    ax.bar_label(container, padding=3, fmt='%.0f')
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion : Phones with highest Sales and fasteners with lowest sale

# Sales and profit by customer segment

```python
sales_profit_by_segment=df.groupby('Segment').agg({'Sales':'sum', 'Profit':'sum'}).reset_index()
sales_profit_by_segment
```

```python
melted = sales_profit_by_segment.melt(id_vars='Segment', var_name='Metric', value_name='Amount')
melted
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=melted, x='Segment', y='Amount', hue='Metric', palette='magma')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f', padding=3)

plt.title('Sales and Profit by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

# Conclusion : sales and profit in descending order : Consumer , Corporate , Home Office
